```python
# load all cleaned dfs
dummy_2018 = pd.read_csv('dummy_2018.csv')
dummy_2019 = pd.read_csv('dummy_2019.csv')
dummy_2020 = pd.read_csv('dummy_2020.csv')
dummy_2021 = pd.read_csv('dummy_2021.csv')
dummy_2022 = pd.read_csv('dummy_2022.csv')
```

```python
def check_classimbalances(df, year): 
    num_approved = (df['action_taken'] == 1).sum()
    num_denied = (df['action_taken'] == 0).sum()
    class_imb = num_denied/num_approved
    print(f'Class imbalance ratio {year}:', class_imb)
```

```python
check_classimbalances(dummy_2018, '2018')
check_classimbalances(dummy_2019, '2019')
check_classimbalances(dummy_2020, '2020')
check_classimbalances(dummy_2021, '2021')
check_classimbalances(dummy_2022, '2022')
```
