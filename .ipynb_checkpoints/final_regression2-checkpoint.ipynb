```python
#imports
import pandas as pd
import numpy as np
import scipy.stats as stats
import statsmodels.api as sm
import matplotlib.pyplot as plt
from sklearn.linear_model import Lasso, LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split, GridSearchCV, KFold, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score, mean_squared_error, confusion_matrix, ConfusionMatrixDisplay, classification_report, roc_curve, roc_auc_score
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
#split training and testing
def split_test_training(df):
    X_train, X_test, y_train, y_test = train_test_split(df.drop('action_taken', axis=1), df['action_taken'], test_size=0.2, random_state=42)
    
    # Fit the scaler on the training data and transform both the training and test data
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)
    
    # Add a constant term for the intercept after scaling
    X_train_scaled = sm.add_constant(X_train_scaled, has_constant='add')
    X_test_scaled = sm.add_constant(X_test_scaled, has_constant='add')
    
    feature_names = df.drop('rate_spread', axis=1).columns.tolist()
    feature_names = ['const'] + feature_names
    X_train_scaled = pd.DataFrame(data=X_train_scaled, columns=feature_names)
    X_test_scaled = pd.DataFrame(data=X_test_scaled, columns=feature_names)

    return X_train_scaled, X_test_scaled, y_train, y_test
```

```python
def logistic_regression(X_train, y_train):
    log_reg = sm.Logit(y_train, X_train)
    log_model = log_reg.fit()
    return log_model
```

```python
#print
def print_summary(log_model, year):
    summary = log_model.summary().as_text()
    lines = summary.split('\n')
    fig, ax = plt.subplots(figsize=(12, len(lines) * 0.3))
    ax.axis('off')
    for i, line in enumerate(lines):
        ax.text(0, len(lines) - i - 1, line, fontsize=10, fontfamily='monospace')
    plt.savefig(f'summary_{year}', bbox_inches='tight')
    plt.close()
```

```python
#print
def classificationreport(log_model, X_test, y_test, year):
    y_pred_prob = log_model.predict(X_test)
    y_pred = np.where(y_pred_prob > 0.5, 1, 0)
    class_report = classification_report(y_test, y_pred)
    
    # Plotting the classification report
    plt.figure(figsize=(10, 10))
    plt.text(0.01, 1.25, 'Classification Report ' + year, {'fontsize': 16}, fontproperties='monospace')
    plt.text(0.01, 0.05, class_report, {'fontsize': 10}, fontproperties='monospace')
    plt.axis('off')
    
    # Save the plot as a PNG file
    plt.savefig(f'classification_report_2_{year}.png')
    plt.close
```

```python
#print
def print_cm_model(log_model, X_test, y_test, year):
    y_pred_prob = log_model.predict(X_test)
    y_pred = np.where(y_pred_prob > 0.5, 1, 0)
    cm = confusion_matrix(y_test, y_pred)
    disp = ConfusionMatrixDisplay.from_predictions(y_test, y_pred, cmap=plt.cm.Blues, normalize='all')
    disp.ax_.set_title(f'Confusion Matrix for {year} Model')
    
    # Save the plot as a PNG file
    plt.savefig(f'confusion_matrix_2_{year}.png')
    plt.close()
```

```python
#print
def plot_roc_curve(log_model, X_test, y_test, year):
    y_prob = log_model.predict(X_test)
    fpr, tpr, thresholds = roc_curve(y_test, y_prob)
    auc_score = roc_auc_score(y_test, y_prob)

    plt.figure()
    plt.plot(fpr, tpr, color='blue', label='ROC curve (area = %0.2f)' % auc_score)
    plt.plot([0, 1], [0, 1], color='grey', linestyle='--')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title(f'Receiver Operating Characteristic (ROC) Curve {year}')
    plt.legend(loc="lower right")
    plt.savefig(f'roc_curve_2_{year}.png')
    plt.close()
```

```python
#2018
dummy_2018 = pd.read_csv('dummy_2018.csv')
X_train_2018, X_test_2018, y_train_2018, y_test_2018 = split_test_training(dummy_2018)
model_2018 = logistic_regression(X_train_2018, y_train_2018)
print_summary(model_2018, '2018')
classificationreport(model_2018, X_test_2018, y_test_2018, '2018')
print_cm_model(model_2018, X_test_2018, y_test_2018, '2018')
plot_roc_curve(model_2018, X_test_2018, y_test_2018, '2018')
```
